```python
# Q1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Q2: Load dataset
df = pd.read_csv('train.csv')
df.head()

# Q3: Column names
df.columns

# Q4: Data types
df.dtypes

# Q5: Shape
df.shape

# Q6: Summary statistics
df.describe()

# Q7: Missing values
df.isnull().sum()

# Q9: Bar charts
sns.countplot(x='Survived', data=df)
plt.show()

sns.countplot(x='Pclass', data=df)
plt.show()

sns.countplot(x='Sex', data=df)
plt.show()

sns.countplot(x='Embarked', data=df)
plt.show()

# Q10: Histogram of Age
sns.histplot(data=df, x='Age', hue='Survived')
plt.show()

# Q11–Q13: Fill missing values
df['Embarked'].fillna('S', inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Q14–Q18: Feature Engineering
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = 1
df.loc[df['FamilySize'] > 1, 'IsAlone'] = 0

# Q19–Q20: Drop unnecessary columns
df.drop(['PassengerId','Name','Ticket','Cabin'], axis=1, inplace=True)

# Convert categorical to numeric
df = pd.get_dummies(df, drop_first=True)

# Q21: Top 5
df.head()

# Q22: Prepare X and Y
X = df.drop('Survived', axis=1)
Y = df['Survived']

# Q23: Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Q25–Q26: Create and train default model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, Y_train)

# Q27: Print tree

plot_tree(model, filled=True)
plt.show()

# Q28–Q32: Evaluation
Y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(Y_test, Y_pred))
print('Confusion Matrix:\n', confusion_matrix(Y_test, Y_pred))
print('Classification Report:\n', classification_report(Y_test, Y_pred))

# Q33: Define max_depth values
param_grid = {'max_depth': [1,2,3,4,5,6,7,8,9,20,22]}

# Q34: Create GridSearch (5-fold CV)
grid = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5)

# Q35: Fit GridSearch
grid.fit(X_train, Y_train)

# Q36: Best max_depth
print('Best max_depth:', grid.best_params_)

# Q37: Best cross-validation score
print('Best CV Score:', grid.best_score_)

# Q38: Store results
results = pd.DataFrame(grid.cv_results_)
results

# Q39: Plot max_depth vs accuracy
plt.plot(results['param_max_depth'], results['mean_test_score'])
plt.xlabel('max_depth')
plt.ylabel('Mean Accuracy')
plt.show()

# Q40: Create optimal model
best_depth = grid.best_params_['max_depth']
optimal_model = DecisionTreeClassifier(max_depth=best_depth, random_state=42)

# Q41: Train optimal model
optimal_model.fit(X_train, Y_train)

# Q42: Print optimal model
print(optimal_model)
plot_tree(optimal_model, filled=True)
plt.show()

# Q43: Make predictions
Y_pred_opt = optimal_model.predict(X_test)

# Q44: Classification report
print(classification_report(Y_test, Y_pred_opt))

# Q45: Confusion matrix
print(confusion_matrix(Y_test, Y_pred_opt))

# Q46–Q47: Accuracy
print('Optimal Model Accuracy:', accuracy_score(Y_test, Y_pred_opt))

```

```python
# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 2. Import 20news group dataset from scikit learn datasets
from sklearn.datasets import fetch_20newsgroups

# 3. Load 20news group train subset
newsgroups_train = fetch_20newsgroups(subset='train')

# 4. Load 20 news group test subset
newsgroups_test = fetch_20newsgroups(subset='test')

# 5. Print all target labels
print(newsgroups_train.target_names)

# 6. Prepare subset of categories alt.atheism, com.graphics, sci.space
categories = ['alt.atheism', 'comp.graphics', 'sci.space']

# 7. Load 20 news group train subset with three categories
train_subset = fetch_20newsgroups(subset='train', categories=categories)

# 8. Load 20 news group test subset with three categories
test_subset = fetch_20newsgroups(subset='test', categories=categories)

# 9. Print new training set target names(Labels)
print(train_subset.target_names)

# 10. Print news training data of 5th article
print(train_subset.data[1])

# 11. Print shape of data and targets
print("Data length:", len(train_subset.data))
print("Targets shape:", train_subset.target.shape)

# 12. Print training set filenames
print(train_subset.filenames)

# 13. By using Count vectorizer train data into numerical format
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(train_subset.data)

# 14. Use bernoullNB for training
from sklearn.naive_bayes import BernoulliNB
clf_bnb = BernoulliNB()
clf_bnb.fit(X_train_counts, train_subset.target)

# 15. By using countvectorizer convert test data into numeric format
X_test_counts = count_vect.transform(test_subset.data)

# 16. Predict target labels for testing set
y_pred_bnb = clf_bnb.predict(X_test_counts)

# 17. Find accuracy score on test set
from sklearn.metrics import accuracy_score
print("BernoulliNB Accuracy:", accuracy_score(test_subset.target, y_pred_bnb))

# 18. Used Tfidfvectorizer instead of count vectorizer and use multinomial NB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
tfidf_vect = TfidfVectorizer()
X_train_tfidf = tfidf_vect.fit_transform(train_subset.data)
clf_mnb = MultinomialNB()
clf_mnb.fit(X_train_tfidf, train_subset.target)

# 19. Find test set accuracy
X_test_tfidf = tfidf_vect.transform(test_subset.data)
y_pred_mnb = clf_mnb.predict(X_test_tfidf)
print("MultinomialNB Accuracy:", accuracy_score(test_subset.target, y_pred_mnb))

# 20. Try with avoiding stopwords and repet the same
tfidf_sw = TfidfVectorizer(stop_words='english')
X_train_sw = tfidf_sw.fit_transform(train_subset.data)
clf_mnb_sw = MultinomialNB()
clf_mnb_sw.fit(X_train_sw, train_subset.target)
X_test_sw = tfidf_sw.transform(test_subset.data)
y_pred_sw = clf_mnb_sw.predict(X_test_sw)
print("MultinomialNB (No Stopwords) Accuracy:", accuracy_score(test_subset.target, y_pred_sw))

```
